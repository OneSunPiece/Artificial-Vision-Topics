## Negativo de una imagen

Este punto consiste en tomar la imagen "bacteria.jpeg" y convertirla a su negativo. El negativo de una imagen se obtiene restando a cada pixel el valor 255, pero aquí está la novedad: vamos a realizar el negativo no en escala de grises, sino en cada canal de color por separado. Es decir, si un pixel tiene el valor (r, g, b), el negativo de ese pixel será (255-r, 255-g, 255-b).

### Imagen original

```python
bacterias = Image("res/Actividad1/bacteria.jpeg")
bacterias
```

```python
bacterias = cv2.cvtColor(cv2.imread("res/Actividad1/bacteria.jpeg"),cv2.COLOR_BGR2RGB)
# creo una imagen negra del mismo tamaño de la anterior
bacteriasNegativo = np.zeros(bacterias.shape,dtype=np.uint8)
colors = ('r','g','b')

for channel in range(len(colors)):
    bacteriasNegativo[:,:,channel] = 255 - bacterias[:,:,channel]

fig, ( (ax1,ax2) , (ax3,ax4) ) = plt.subplots(2, 2, figsize=(15,15))

ax1.imshow(bacterias)
ax1.set_title('Imagen Original')

for i,col  in enumerate(colors):
    temp = bacterias[:,:,i].ravel()
    ax2.hist(temp,histtype = 'step',bins=255,range=(0.0,255.0),density=True,color=colors[i])

ax3.imshow(bacteriasNegativo)
ax3.set_title('Imagen Negativa')

for i,col  in enumerate(colors):
    temp = bacteriasNegativo[:,:,i].ravel()
    ax4.hist(temp,histtype = 'step',bins=255,range=(0.0,255.0),density=True,color=colors[i])
```

## Encoger el histograma

Disminuya el histograma de la imagen "teatro.jpg" en un 50%. El resultado esperado son cuatro imágenes: la imagen original, el histograma original, la imagen modificada y el histograma modificado.

```python
Image("res/Actividad2/teatro.jpg")
```

```python
teatro = cv2.cvtColor(cv2.imread("res/Actividad2/teatro.jpg"),cv2.COLOR_BGR2RGB)

canvas = np.zeros(teatro.shape,dtype=np.uint8)

canvas[:,:] = teatro[:,:]* 0.5

fig, ( (ax1,ax2) , (ax3,ax4) ) = plt.subplots(2, 2, figsize=(15,15))

ax1.imshow(teatro)
ax1.set_title('Imagen Original')

for i,col  in enumerate(colors):
    temp = teatro[:,:,i].ravel()
    ax2.hist(temp,histtype = 'step',bins=255,range=(0.0,255.0),density=True,color=colors[i])

ax3.imshow(canvas)
ax3.set_title('Imagen Opaca')

for i,col  in enumerate(colors):
    temp = canvas[:,:,i].ravel()
    ax4.hist(temp,histtype = 'step',bins=255,range=(0.0,255.0),density=True,color=colors[i])
```
